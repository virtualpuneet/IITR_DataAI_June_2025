```python
# To convert number in your Base to your Base 10
m = int(input ('your base:'))
n = int(input ('your base number:'))
fin_out=0
#split the number as seperate digits and multiply base to the power 0, 1, 2 and ...
str_num = str(n)
str_len = len(str_num)
for i in str_num:
    str_len=str_len-1
    out=int(int(i)*(m**str_len))
    fin_out=fin_out+out
print("Base 10 equivalent is: ", fin_out)
```

```python
# Program to convert any base to any base

# Function to convert number in your Base to your Base 10 (Decimal)
def base_to_dec(m,n):
    #m = int(input ('your base:'))
    #n = int(input ('your base number:'))
    fin_out=0
    #split the number as seperate digits and multiply base to the power 0, 1, 2 and ...
    str_num = str(n)
    str_len = len(str_num)
    for i in str_num:
        str_len=str_len-1
        out=int(int(i)*(m**str_len))
        fin_out=fin_out+out
    return (fin_out)


# Function to convert Base 10 (Decimal) number to your Base
def dec_to_base(m,n):
    #m = int(input ('your base:'))
    #n = int(input ('Base 10 number:'))
    fin_out=str()
    while n//m != 0:
        rem=n%m
        #print (rem)
        out=str(rem)
        fin_out = out+fin_out
        #print(fin_out)
        n= n//m
    #print (str(n%m))
    out=str(n%m)
    fin_out = out+fin_out
    return(int(fin_out))

a = int (input ('enter your current base: '))
input_num = int (input ('enter your current number: '))
b = int (input ('enter the desired base: '))

b2d=base_to_dec(a,input_num)
d2b=dec_to_base(b,b2d)
print ("number in your desired based is:" , d2b)
```

# Calculate Interest

```python
def calculate_interest(principal, rate, time):
    interest = (principal * rate * time)/100
    return (interest)
print ("The interest is: ", calculate_interest(500,10,0))
```

# Calculate Hypotense

```python
def calculate_hypotenuse(base, height):
    hypotenuse = (base**2 + height**2) ** 0.5
    return (hypotenuse)
print ("The hypotenuse is: ", calculate_hypotenuse(3,4))
print ("The hypotenuse is: ", calculate_hypotenuse(5,12))
print ("The hypotenuse is: ", calculate_hypotenuse(0,0))
```

# Find Distance 2D

```python
def calculate_hypotenuse(base, height):
    hypotenuse = (base**2 + height**2) ** 0.5
    return (hypotenuse)

def find_distance_2d(x1, y1, x2, y2):
    base = x2-x1
    height = y2-y1
    return(calculate_hypotenuse (base, height))
print (find_distance_2d(0, 0, 3, 4))
print(find_distance_2d(1, 2, 4, 6))
    
```

# Find Distance 3D


```python
import math
def calculate_distance_3d(x1, y1, z1, x2, y2, z2):
    base_a=x2-x1 # base rectangle length
    base_b=y2-y1 # base rectangle width
    base_c = math.sqrt(base_a**2+base_b**2) # hypothenues of base rectangle
    h = z2-z1
    return(math.sqrt(h**2+base_c**2))
calculate_distance_3d(1, 2, 3, 4, 6, 9)
```
