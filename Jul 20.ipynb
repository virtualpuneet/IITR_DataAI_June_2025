```python
def convert_base10_to_my_base(number, symbols):
    my_base = len(symbols)
    print(f"My Base: {my_base}")
    result = []
    mult = 1
    while number > 0:
        rem = number % my_base
        number = number // my_base
        print(f"Remainder: {rem}")
        sym = symbols[rem]
        print(sym)
        result.append(sym)
    print(f"Before reversing: {result}")
    new_rev = result[::-1]
    print(f"After reversing: {new_rev}")
    return "".join(new_rev)
```

```python
symbols = list('ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+/')
convert_base10_to_my_base(213987129371238, symbols)
```
